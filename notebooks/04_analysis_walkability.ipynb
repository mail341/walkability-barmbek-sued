# Analysis of insights of the district

```python
import geopandas as gpd
import pandas as pd
```

```python
df= pd.read_csv("../data/mart/walkability_data_mart.csv", sep=",")
```

```python
df.columns
```

```python
# --- KPIs berechnen ---
df["stops_per_1000"] = df["pt_stops"] / df["population"] * 1000
df["crossings_per_1000"] = df["crossings"] / df["population"] * 1000
df["bike_km_per_1000"] = df["bike_km"] / df["population"] * 1000
df["green_per_person"] = df["green_area_m2"] / df["population"]

# --- Prozentwerte ---
df["elderly_share"] = df["elderly"] / df["population"] * 100
df["children_share"] = df["children"] / df["population"] * 100

# --- Runden (für saubere Darstellung) ---
df["stops_per_1000"] = df["stops_per_1000"].round(3)
df["crossings_per_1000"] = df["crossings_per_1000"].round(3)
df["bike_km_per_1000"] = df["bike_km_per_1000"].round(3)
df["green_per_person"] = df["green_per_person"].round(2)
df["elderly_share"] = df["elderly_share"].round(1)
df["children_share"] = df["children_share"].round(1)

# --- Kontrolle ---
df[[
    "stops_per_1000",
    "crossings_per_1000",
    "bike_km_per_1000",
    "green_per_person",
    "elderly_share",
    "children_share"
]]
```

```python

df.to_csv(
    "../data/mart/walkability_data_mart.csv",
    index=False,
    sep=";",              # wichtig für DE Systeme
    decimal=","           # 🔥 entscheidend!
)
```
